```python
#Using DBSCAN for outlier detection
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np

# Standardize the data for DBSCAN
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[numeric_cols])

# Apply DBSCAN to the scaled data
dbscan = DBSCAN(eps=0.5, min_samples=5)
predictions = dbscan.fit_predict(data_scaled)

# Find the indices of the outliers
outliers = np.where(predictions == -1)

# Print the number of outliers detected
print(f"Number of outliers detected by DBSCAN: {len(outliers[0])}")
```

this particular dataset is not suitable for density blased clustering; there are no clear clusters in the data, that is why dbscan is returning all total observation as outliers

I am not confident that the outlier is due to incorrectly entered or measured data.
so i would just replace values that are above a certain upper limit or below a certain lower limit with the upper or lower limit, respectively

Iterating through each column in the DataFrame and calculating the upper and lower limits for each column using the 95th and 5th percentiles. Then, it caps values above the upper limit to the upper limit and values below the lower limit to the lower limit.

```python
# Define the name of the target variable
target_variable = 'diagnosis'

# Capping outliers using percentile
for column in data.columns:
    # Skip the target variable
    if column == target_variable:
        continue
    
    # Calculate the upper and lower limits for each column using the 50th and 5th percentiles
    upper_lim = data[column].quantile(.50)
    lower_lim = data[column].quantile(.05)
    
    # Cap values above the upper limit to the upper limit
    data.loc[(data[column] > upper_lim),column] = upper_lim
    
    # Cap values below the lower limit to the lower limit
    data.loc[(data[column] < lower_lim),column] = lower_lim
```

Model building and training without feature selection

```python
# Define the models
models = [
    ("Logistic Regression", LogisticRegression()),
    ("Decision Tree", DecisionTreeClassifier()),
    ("Random Forest", RandomForestClassifier()),
    ("SVM", SVC()),
    ("Gradient Boosting", GradientBoostingClassifier()),
]
```
